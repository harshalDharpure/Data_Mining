```python
import pandas as pd
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import apriori
```

```python
data=pd.read_csv("restaurants-challenge.csv",header=None)
data.head()
```

```python
transaction=[]
for i in range (0,179730):
    transaction.append([str(data.values[i,j])
                       for j in range (0,3)])
transaction[0:5]
```

```python
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_transaction=te.fit_transform(transaction)
te_transaction
```

```python
te.columns_
```

```python
data1=pd.DataFrame(te_transaction,columns=te.columns_)
data1.head()
```

```python
model = apriori(data1, min_support =0.02, use_colnames = True)
results = association_rules(model, 
                            metric = "confidence", 
                            min_threshold = 0.5)
results.sort_values(by='lift', ascending=False)
```

```python
import matplotlib.pyplot as plt
plt.scatter(results['support'], 
            results['confidence'],
            s = 100,
            color = 'red')
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
for i, label in enumerate(results.index):
    plt.annotate(label, (results['support'][i], 
                         results['confidence'][i]))
```
